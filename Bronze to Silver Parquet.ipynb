```python
from pyspark.sql.functions import col

# Read raw CSV from Bronze Lakehouse (assuming file path is accessible)
df = spark.read.option("header", True).csv("Files/Full_Pipeline/full_pipeline_raw_sales_10k.csv")

df.head(5)
```

```python
# Basic Cleaning (casting, trimming)
df_cleaned = df.withColumn("quantity", col("quantity").cast("int")) \
               .withColumn("unit_price", col("unit_price").cast("double")) \
               .withColumn("total_amount", col("total_amount").cast("double")) \
               .withColumn("sale_date", col("sale_date").cast("date"))

df_cleaned.head(5)
```

```python
# Save as Parquet to Silver Lakehouse
df_cleaned.write.mode("overwrite").format("parquet").save("abfss://BARQ_Intenrship@onelake.dfs.fabric.microsoft.com/Silver_lakehouse.Lakehouse/Files/sales_data/sales_cleaned")
```
