```python
# Welcome to your new notebook
# Type here in the cell editor to add code!


df = spark.read.parquet("Files/sales_data/sales_cleaned")

df.show()
```

```python
from pyspark.sql.functions import row_number
from pyspark.sql.window import Window

dim_customer = df.select("first_name","last_name").distinct().withColumn("Dim_Customer_ID",row_number().over(Window.orderBy("first_name","last_name")))
dim_customer.show()
```

```python
dim_store = df.select("store_city").distinct().withColumn("Dim_Store_ID",row_number().over(Window.orderBy("store_city")))
dim_store.show()
```

```python
dim_product = df.select("product_category","color","brand","unit_price").distinct().withColumn("Dim_Product_ID",row_number().over(Window.orderBy("product_category","color","brand","unit_price")))
dim_product.show()
```

```python
Gold_lakehouse_Path = "abfss://BARQ_Intenrship@onelake.dfs.fabric.microsoft.com/Gold_lakehouse.Lakehouse/Tables"

dim_customer.write.mode("overwrite").format("delta").save(f"{Gold_lakehouse_Path}/DimCustomer")
dim_product.write.mode("overwrite").format("delta").save(f"{Gold_lakehouse_Path}/DimProduct")
dim_store.write.mode("overwrite").format("delta").save(f"{Gold_lakehouse_Path}/DimStore")
```
